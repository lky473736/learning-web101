### 연습문제 2

```python
%%html

<!doctype html5>
<html lang="ko">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>연습문제 2</title>
    </head>
    <body>
        <form action="post">
            <fieldset>
                <legend>납품자 정보</legend>
                <ol>
                    <li>
                        <p>납품자명 : <input type="text" name="name" id="name"></p>
                    </li>
                    <li>
                        <p>email : <input type="email" name="email" id="email" placeholder="answer@naver.com"></p>
                    </li>
                    <li>
                        <p>홈페이지 : <input type="url" name="url" id="url" placeholder="http://"></p>
                    </li>
                    <li>
                        <p>지역 : 
                            <select name="location" id="location">
                                <option value="pusan" selected>부산</option>
                                <option value="seoul">서울</option>
                                <option value="seongnam">성남</option>
                            </select>
                        </p>
                    </li>
                </ol>
            </fieldset>
            <fieldset>
                <legend>납품 정보</legend>
                <ul>
                    <li>
                        <p>
                            상품명 : <input type="text" name="objname" id="objname">
                        </p>
                    </li>
                    <li>
                        <p>
                            납품수량 : <input type="number" name="objnum" id="objnum" placeholder="최소 100">
                        </p>
                    </li>
                    <li>
                        <p>
                            납품등급 : <input type="range" name="grade" id="grade">
                        </p>
                    </li>
                    <li>
                        <p>
                            기타사항 : <textarea name="notice" id="notice" cols="30" rows="10"></textarea>
                        </p>
                    </li>
                </ul>
            </fieldset>

            <button type="submit">send message</button>
        </form>
    </body>
</html>
```
