### HTML 문서 기본 구조

- doctype : 현재 문서의 html 타입 명시  
- html : html 문서의 시작과 끝
- head : 웹페이지 문서의 대한 정보 (metadata)
    - title, style, meta, link, script, base 등을 head에서 정의
    - title : html 문서 제목
- body : 내부에 웹페이지의 실제적인 내용  



```python
%%HTML
<!doctype html5>
<html>
    <head>
        <title>HTML 문서</title>
    </head>
    <body>
        <h1>제목</h1>
        <p>단락</p>
    </body>
</html>
```

-------

### html 문서 요소, 속성

```python
%%HTML
<p class="value">단락</p>
```

줄 하나하나를 요소(element)라고 생각하면 편함  
- 태그 이름 : p
- 속성명 : class
- 속성값 : value
- 내용(content) : 단락

```python
%%HTML
<a href="https://naver.com">링크</a>
```

* 태그 이름 : a
* 속성명 : href
* 속성값 : 링크의 주소
* 내용 : 링크  

각 태그마다 반드시 필요한 속성이나 기능을 추가하는 속성이 정해져 있다.

### html 문서 작성 규칙

- 연속된 공백 또는 줄바꿈은 하나의 공백으로 처리한다
- 여러개의 공백이나 줄바꿈은 다른 특수문자를 사용한다
- 시작 태그와 종료 태그 쌍이 서로 교차되지 않아야 한다

```python
%%html

<!doctype html5>
<html>
    <head>
        <title>공백 처리 예제</title>
    </head>
    <body>
        공         백  
        줄바꿈
        줄바꿈
    </body>
</html>
```

```python
%%html

<!doctype html5>
<html>
    <head>
        <title>잘못된 태그 예제</title>
    </head>
    <body>
        <h1>잘못된<p>태그</h1></p> <!--교차 X-->
    </body>
</html>
```

### 텍스트 요소 태그

#### 제목 태그   
- h태그. 1~6단계의 레벨이 있음 (숫자가 높아질 수록 작아짐)

```python
%%html
<h1>Heading 1</h1>
<h2>Heading 2</h2>
<h3>Heading 3</h3>
<h4>Heading 4</h4>
<h5>Heading 5</h5>
<h6>Heading 6</h6>
```

#### 단락 태그   
- p태그. 문단 전 후에 공간이 띄워짐

```python
%%html
<p>단락</p>
<p>단락2</p>
```

#### 개행 태그, 수평선 태그    
- br태그. 강제로 줄바꿈하기
- hr태그. 수평선

```python
%%html
<p>
    줄나누기
    <br>
    이렇게줄나누기 (개행)
    <hr>
</p>
```

#### 미리 정의된 서식 태그

- pre태그. 텍스트 그대로 화면에 표시할 때 사용

```python
%%html

<pre>
    그    대     로     출         력

    그    대     로 
</pre>
```

### b, i, u, s...

- b태그 : 굵게
- i태그 : 이텔릭
- u태그, ins태그 : 밑줄
- s태그, del태그 : 가운데선
- mark태그 : 하이라이트
- strong태그 : 중요
- code태그 : 코드
- sup태그 : 윗첨자
- sub태그 : 아랫첨자

```python
%%html

<b>굵게</b>
<br>
<i>이텔릭</i>
<br>
<small>작게</small>
<br>
<ins>밑줄</ins>
<br>
<del>가운데 선</del>
<br>
<mark>하이라이트</mark>
<br>
<strong>중요</strong>
<br>
<code>코드</code>
<br>
<em>강조</em>
<br>
임규연<sup>위 첨자</sup>
<br>
임규연<sub>아래 첨자</sub>
```
