### css 개요  
  
- 웹 페이지의 디자인 구성

### css 구성 요소  
  
- 선택자 (selector) : 스타일을 적용할 대상을 지정함
- 속성명 (property != attribute) : 속성의 이름
- 속성값 (value) : 속성에 적용된 값

```python
%%html
p {
    color : purple;
    text-align : center;
}
```

- 선택자 : p. 모든 p태그에 저 스타일이 적용된다
- 속성명 : color, text-align
- 속성값 : purple, center

--------

### css 적용
- 인라인 스타일 : html 요소의 open tag에 스타일 속성 지정
- 내부 스타일 : html의 head 태그 안에 스타일 속성 지정
- 외부 스타일 : html 문서 별개의 파일로 스타일 속성 지정 (head 태그 안에 link를 이용해 join)
    - <code>link rel = "stylesheet" href = "css파일경로"</code>

### css 선택자

![스크린샷 2024-02-17 오후 8.05.11.png](<attachment:스크린샷 2024-02-17 오후 8.05.11.png>)

- 내용 안에는 속성명과 속성값이 pair 형태로 이루어짐
- 모든 요소에 스타일을 부여할 때는 * 기호를 이용 
- 특정 태그에 스타일을 부여할 때는 태그명을 이용
- 특정 id 태그에만 스타일을 부여할 때는 # 이용
- 특정 class 태그에만 스타일을 부여할 때는 . 이용 
- 특정 선택자 안에서 특정 속성이 있을 때 -> 속성 선택자를 이용

```python
%%html
<!DOCTYPE html>
<html>
  <head>
    <style>
      p { /* p태그 전부 */
        background-color: yellow;
      }
      #id1 { /* id1이라는 id를 가진 태그 전부 */
        color: yellow;
        background-color: black;
      }
      .class1 { /* class1이라는 class를 가진 태그 전부 */
        color: blue;
        background-color: yellow;
      }
      p.class1 { /* class1이라는 class를 가진 태그 중 p태그 전부 */
        color: red;
        background-color: navy;
      }
      h1 { /* h1태그 전부 */
        color: green;
      }
      h1[text] { /* h1태그인데 속성으로 text를 가지고 있는 태그 전부 */
        color: purple;
      }
      input[type="password"] { /* input태그인데 속성으로 type이고, 속성값이 password인 태그 전부 */
        background: yellow;
      }
    </style>
  </head>
  <body>
    <p>p</p>
    <p id="id1">id1인 p</p>
    <p class="class1">class1인 p</p>
    <p class="class2">class2인 p</p>
    <h1>h1</h1>
    <h1 text="h1_text">text 속성 h1</h1>
    <form>
      <input type="password">
    </form>
  </body>
</html>
```

### 반응 선택자

![스크린샷 2024-02-17 오후 8.29.48.png](<attachment:스크린샷 2024-02-17 오후 8.29.48.png>)

- a태그는 방문했냐 안했냐로 구분 
    - 방문하지 않은 링크 : link
    - 방문한 링크 : visited
- 마우스의 반응에 의해서도 스타일을 부여할 수 있음
    - 클릭 : active
    - 올려놓기 : hover

```python
%%html

<!DOCTYPE html>
<html>
  <head>
    <style>
      a:link { color: blue; } /* 방문하지 않았다면 */
      a:visited { color: purple; } /* 방문한 적 있다면 */
      a:hover { color: red; } /* 마우스를 올려놓을 때 */
      a:active { color: black; } /* 클릭할 때 */
  
      span.c1:hover { background-color: yellow; } /* c1 클래스에 있는 모든 span 요소에 마우스를 올려놓을 때 */
    </style>
  </head>
  <body>
    <a href="http://www.google.com" target="_blank">구글 링크</a>
    <span class="c1">
       span영역
    </div>
  </body>
</html>
```

```python
%%html

<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>반응 선택자 연습</title>
    <style>
        button {
            background-color : blue;
            color : white;
            font-size : 16px;
        }

        button:hover {
            background-color : red;
        }

        button:active {
            background-color: yellow;
            color : black;
        }
    </style>
</head>
<body>
    <button>버튼</button> 
</body>
</html>
```

### 구조 선택자  
![스크린샷 2024-02-17 오후 8.42.16.png](<attachment:스크린샷 2024-02-17 오후 8.42.16.png>)

- css는 태그를 부모와 자식 간 상속이 가능한 스크립트이다
- 따라서 구조적으로 선택자를 구성할 수 있다

```python
%%html
<!DOCTYPE html>
<html>
<head>
    <style>
        p { color : blue; background-color: white;}
        /* nth-child() : 특정 태그 사이에서 몇번째 태그 */
        p:nth-child(1) { font-size: 400%; } /* 첫번째 p 태그 */
        p:nth-child(2) { font-size: 350%; } /* 두번째 p 태그 */
        p:nth-child(3) { font-size: 300%; } /* 세번째 p 태그 */
        p:nth-child(4) { font-size: 250%; } /* 네번째 p 태그 */
        p:nth-child(5) { font-size: 200%; } /* 다섯번째 p 태그 */
        p:nth-child(6) { font-size: 150%; } /* 여섯번째 p 태그 */
        p:nth-child(7) { font-size: 100%; } /* 일곱번째 p 태그 */
    </style>
</head>
<body>
    <p>lky473736</p>
    <p>lky473736</p>
    <p>lky473736</p>
    <p>lky473736</p>
    <p>lky473736</p>
    <p>lky473736</p>
    <p>lky473736</p>
</body>
</html>
```

### 상태 선택자 (form의 상태를 선택하여 사용)

![스크린샷 2024-02-17 오후 9.22.00.png](<attachment:스크린샷 2024-02-17 오후 9.22.00.png>)
- enabled : 사용 가능한 (입력 가능한) form (form의 기본이 enabled)
- disabled : 사용 불가능한 (입력이 불가능한) form
- focus : 내가 입력하려고 form을 누르는 순간 (키보드 깜빡이가 열리는 순간)

```python
%%html
<!DOCTYPE html>
<html>
<head>
    <style>
        input:enabled { background-color: skyblue; } /*입력 가능한 input*/
        input:disabled { background-color: darkgray; } /*입력이 불가능한 input*/
        input:focus { background-color: yellow; } /*입력가능한 input & 내가 form을 누르는 순간*/
    </style>
</head>
<body>
    <input>
    <input disabled="disabled"/> <!--입력이 불가능하게 된다-->
    <input>
</body>
</html>
```

### 선택자 조합

![스크린샷 2024-02-17 오후 9.28.45.png](<attachment:스크린샷 2024-02-17 오후 9.28.45.png>)
- A B : A 안에 있는 B
- A > B : A의 open tag에 가장 가까운 B (A B + 가장 첫번째 B)
- A + B : A 바로 다음의 B 
- A ~ B : A 다음에 인접한 모든 B
- A B : A 그리고 B 

```python
%%html
<!DOCTYPE html>
<html>
<head>
    <style>
        div h1 { color: red; } /*div태그 안 h1*/
        h1 + h2 { color: green; } /* h1태그 바로 다음 h2태그에*/
        h2 ~ h4 { background-color: yellow; } /*h2태그와 인접한 h4*/
        h5, h6 { font-size: 200% } /*h5태그와 h6태그*/
    </style>
</head>
<body>
    <div>
        <h1>Heading 1</h1>
        <h2>Heading 2</h2>
        <h2>another Heading 2</h2>
        <h3>Heading 3</h3>
        <h4>Heading 4</h4>
        <h4>another Heading 4</h4>
        <h5>Heading 5</h5>
        <h6>Heading 6</h6>
    </div>
</body>
</html>
```
